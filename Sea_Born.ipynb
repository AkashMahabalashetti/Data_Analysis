```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('C:\\Users\\Hp\\OneDrive\\Desktop\\Data_Analyst\\Superstore.xlsx')
df
```

## relplot

```python
sns.relplot(data=df, x="Sub-Category", y="Sales", hue="Category")
```

```python
sns.relplot(data=df, x="Sub-Category", y="Sales", hue="Category",col='Region',col_wrap=2)
```

```python
sns.relplot(data=df, x="Sub-Category", y="Sales", row ="Region",col='Category',hue='Category')
```

```python
sns.relplot(data=df, x="Sub-Category", y="Sales", hue="Category",kind='line')
```

## lineplot

```python
sns.lineplot(data=df, x="Sales", y="Profit",hue='Sub-Category')
```

## catplot

```python
sns.catplot(data=df, x="Sales", y="Category")
```

```python
sns.catplot(data=df, x="Sales", y="Category",kind='box')
```

```python
sns.catplot(data=df, x="Sales", y="Category",kind='violin',hue='Category')
```

## barplot

```python
sns.barplot(data=df,x='Sub-Category',y='Sales',hue='Category')
```

## countplot

```python
sns.countplot(df,x='Sub-Category')
```

## pairplot

```python
sns.pairplot(df,hue='Category')
```
